```python
import pandas as pd
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
import math
```

```python
file_raw_data = "C:\\Users\Allen\Python Workspace\data-set\heart_disease_.csv"
raw_data = pd.read_csv(file_raw_data)
raw_data
```

```python
x = raw_data.iloc[:,0:22]
x
```

```python
num_features = len(x.columns)

print("Number of Features: {}".format(num_features))
```

```python
columns = []

for i in range(num_features):
    columns.append("x{}".format(i))

x.columns = columns

x
```

```python
x_mean = x.mean(numeric_only=True)
x_std = x.std(numeric_only=True)
x_standardized = (x - x_mean)/x_std
x_standardized
```

```python
x_normalized = (x - x.min()) / (x.max() - x.min())

x_normalized
```

```python
y = raw_data['Sex'].replace([0.0], 0).replace([1.0], 1)

y = y.values

y
```

```python
df = x_normalized.copy()
df['y'] = y

df
```

```python
def partition_dataset(df, num_a=50, num_b=50, val_a=1, val_b=0):
    df_a = df[df.iloc[:,18] == val_a].sample(num_a)
    df_b = df[df.iloc[:,18] == val_b].sample(num_b)
    
    df.drop(df_a.index, inplace=True)
    df.drop(df_b.index, inplace=True)
    
    frames = [df_a, df_b]
    df_validation = pd.concat(frames)
    
    return df, df_validation

training, validation = partition_dataset(df, num_a=50, num_b=50)
```

```python
training
```

```python
validation
```

```python
x_training = training.iloc[:,:18].values
x_validation = validation.iloc[:,:18].values

y_training = training['y'].values
y_validation = validation['y'].values

y_validation
```

```python
model = KNeighborsClassifier(n_neighbors=10)

model.fit(x_training, y_training)

predictions = model.predict(x_validation)

predictions
```

```python
cm = confusion_matrix(y_validation, predictions)
tp = cm[0][0]
tn = cm[1][1]
fp = cm[0][1]
fn = cm[1][0]

mcc = ((tn * tp) - (fn * fp)) / math.sqrt((tp + fp) * (tp + fn) * (tn + fp) * (tn + fn))
f1 = tp / (tp + (0.5 * (fp + fn)))

print(classification_report(y_validation, predictions))
print("MCC: {}".format(mcc))
print("F1: {}".format(f1))
```

```python
model = LogisticRegression(random_state=0)

model.fit(x_training, y_training)

predictions = model.predict(x_validation)
```

```python
cm = confusion_matrix(y_validation, predictions)
tp = cm[0][0]
tn = cm[1][1]
fp = cm[0][1]
fn = cm[1][0]

mcc = ((tn * tp) - (fn * fp)) / math.sqrt((tp + fp) * (tp + fn) * (tn + fp) * (tn + fn))
f1 = tp / (tp + (0.5 * (fp + fn)))

print(classification_report(y_validation, predictions))
print("MCC: {}".format(mcc))
print("F1: {}".format(f1))
```
