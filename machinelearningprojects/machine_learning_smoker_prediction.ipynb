```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import confusion_matrix, accuracy_score
from matplotlib.colors import ListedColormap
```

```python
dataset = pd.read_csv('./medinsur-1.csv')
dataset.head()
```

```python
dataset.shape
```

```python
mmf = dataset.loc[dataset['sex'] == 'male']
mmf
```

```python
fmf = dataset.loc[dataset['sex'] == 'female']
fmf
```

```python
mmf = mmf.dropna()
fmf = fmf.dropna()
```

```python
X1 = fmf.iloc[:, [0, 2, 3]].values
y1 = fmf.iloc[:, [4]].values

fmf.head()
```

```python
X2 = mmf.iloc[:, [0, 2, 3]].values
y2 = mmf.iloc[:, [4]].values

mmf.head()
```

```python
X1_train, X1_test, y1_train, y1_test = train_test_split(X1
                                                   , y1
                                                   , test_size=0.25
                                                   , random_state=1)
```

```python
X2_train, X2_test, y2_train, y2_test = train_test_split(X2
                                                   , y2
                                                   , test_size=0.25
                                                   , random_state=1)
```

```python
sc1 = StandardScaler()
X1_train = sc1.fit_transform(X1_train)
X1_test = sc1.transform(X1_test)
sc2 = StandardScaler()
X2_train = sc2.fit_transform(X2_train)
X2_test = sc2.transform(X2_test)
```

```python
fem = LogisticRegression()
fem.fit(X1_train, y1_train)
```

```python
mal = LogisticRegression()
mal.fit(X2_train, y2_train)
```

```python
y1_pred = fem.predict(X1_test)
print(np.concatenate((y1_pred.reshape(len(y1_pred),1)
                                    , y1_test.reshape(len(y1_test), 1))
                                                    , 1))
```

```python
y2_pred = mal.predict(X2_test)
print(np.concatenate((y2_pred.reshape(len(y2_pred),1)
                                    , y2_test.reshape(len(y2_test), 1))
                                                    , 1))
```

```python
cm1 = confusion_matrix(y1_test, y1_pred)
print(cm1)

accuracy_score(y1_test, y1_pred)
```

```python
cm2 = confusion_matrix(y2_test, y2_pred)
print(cm2)

accuracy_score(y2_test, y2_pred)
```

```python
from tkinter import *
from tkinter import messagebox


mainFrame = Tk()
mainFrame.title("Medical Insurance Computer")
mainFrame.geometry('500x400')

# Create Frames
frameFirst = Frame(mainFrame)
frameFirst.grid(row=0, column=0)
frameSecond = Frame(mainFrame)
frameSecond.grid(row=0, column=1)
frameThird = Frame(mainFrame)
frameThird.grid(row=0, column=2)

# Set Values (Assuming dataset is defined)
age_min = 18
age_max = 65

bmi_min = 18.5
bmi_max = 24.9

children_min = 0
children_max = 5

# Drop Down Gender
chooseGender = [
    "Male",
    "Female"
]

# DataType of Gender Text
clickedGender = StringVar()
clickedGender.set("Male")

# DataType of Name Text
clickedName = StringVar()
clickedName.set("")

# DataType of Age Text
clickedAge = StringVar()
clickedAge.set("")

# DataType of Weight Text
clickedWeight = StringVar()
clickedWeight.set("")

# DataType of Height Text
clickedHeight = StringVar()
clickedHeight.set("")

# DataType of Children Text
clickedChildren = StringVar()
clickedChildren.set("")

# Visible UI
# Input Areas
# Show Name Input Area
labelName = Label(frameFirst, text="Name")
labelName.grid(row=0, column=0, padx=5, pady=5)
entryName = Entry(frameFirst, textvariable=clickedName)
entryName.grid(row=0, column=1, padx=5, pady=5)

# Show Dropdown Gender Menu
labelGender = Label(frameFirst, text="Gender")
labelGender.grid(row=1, column=0, padx=5, pady=5)
dropGender = OptionMenu(frameFirst, clickedGender, *chooseGender)
dropGender.grid(row=1, column=1, padx=5, pady=5)

# Show Age Input Area
labelAge = Label(frameFirst, text="Age")
labelAge.grid(row=2, column=0, padx=5, pady=5)
entryAge = Entry(frameFirst, textvariable=clickedAge)
entryAge.grid(row=2, column=1, padx=5, pady=5)

# Show Weight Input Area
labelWeight = Label(frameFirst, text="Weight")
labelWeight.grid(row=3, column=0, padx=5, pady=5)
entryWeight = Entry(frameFirst, textvariable=clickedWeight)
entryWeight.grid(row=3, column=1, padx=5, pady=5)

# Show Height Input Area
labelHeight = Label(frameFirst, text="Height")
labelHeight.grid(row=4, column=0, padx=5, pady=5)
entryHeight = Entry(frameFirst, textvariable=clickedHeight)
entryHeight.grid(row=4, column=1, padx=5, pady=5)

# Show Children Input Area
labelChildren = Label(frameFirst, text="Children")
labelChildren.grid(row=5, column=0, padx=5, pady=5)
entryChildren = Entry(frameFirst, textvariable=clickedChildren)
entryChildren.grid(row=5, column=1, padx=5, pady=5)

def show_Results():
    # Output Areas
    # Show Name Output Area
    labelName = Label(frameSecond, text="Name")
    labelName.grid(row=0, column=0, padx=5, pady=5)
    OutputName = Label(frameSecond, text=clickedName.get())
    OutputName.grid(row=0, column=1, padx=5, pady=5)

    # Show Gender Output Menu
    labelGender = Label(frameSecond, text="Gender")
    labelGender.grid(row=1, column=0, padx=5, pady=5)
    OutputGender = Label(frameSecond, text=clickedGender.get())
    OutputGender.grid(row=1, column=1, padx=5, pady=5)

    # Show Age Output Area
    labelAge = Label(frameSecond, text="Age")
    labelAge.grid(row=2, column=0, padx=5, pady=5)
    OutputAge = Label(frameSecond, text=clickedAge.get())
    OutputAge.grid(row=2, column=1, padx=5, pady=5)

    # Show Weight Output Area
    labelWeight = Label(frameSecond, text="Weight(kg)")
    labelWeight.grid(row=3, column=0, padx=5, pady=5)
    OutputWeight = Label(frameSecond, text=clickedWeight.get())
    OutputWeight.grid(row=3, column=1, padx=5, pady=5)

    # Show Height Output Area
    labelHeight = Label(frameSecond, text="Height(cm)")
    labelHeight.grid(row=4, column=0, padx=5, pady=5)
    OutputHeight = Label(frameSecond, text=clickedHeight.get())
    OutputHeight.grid(row=4, column=1, padx=5, pady=5)

    # Show BMI Output Area
    labelBMI = Label(frameSecond, text="BMI")
    labelBMI.grid(row=5, column=0, padx=5, pady=5)
    OutputBMI = Label(frameSecond, text=calculate_bmi(clickedWeight.get(), clickedHeight.get()))
    OutputBMI.grid(row=5, column=1, padx=5, pady=5)

    # Show Children Output Area
    labelChildren = Label(frameSecond, text="Children")
    labelChildren.grid(row=6, column=0, padx=5, pady=5)
    OutputChildren = Label(frameSecond, text=clickedChildren.get())
    OutputChildren.grid(row=6, column=1, padx=5, pady=5)

    # Show Fee Output Area
    user_input = np.array([[float(clickedAge.get()), calculate_bmi(clickedWeight.get(), clickedHeight.get()), float(clickedChildren.get())]])
    labelFee = Label(frameThird, text="Fee")
    labelFee.grid(row=0, column=0, padx=5, pady=5)
    OutputFee = Label(frameThird, text=charges(nsmokermod.intercept, nsmokermod.coef, nsmokermod, user_input))
    OutputFee.grid(row=0, column=1, padx=5, pady=5)

# Calculations
def calculate_bmi(weight, height):
    try:
        weight = float(weight)
        height = float(height)
        height_in_meters = height / 100  # Convert height from cm to meters
        bmi = weight / (height_in_meters ** 2)
        bmi = round(bmi, 2)
        return bmi
    except ValueError:
        return "Invalid input"

# Display Results Button
btn_GetResults = Button(frameFirst, text='Calculate', command=validate_input)
btn_GetResults.grid(row=6, column=0, padx=5, pady=5)

# Value Validation
def validate_input():
    validation_errors = []

    # Validate Age
    try:
        age = float(clickedAge.get())
        if age < age_min or age > age_max:
            validation_errors.append("Age must be between {} and {}.".format(age_min, age_max))
    except ValueError:
        validation_errors.append("Invalid age input.")

    # Validate Weight
    try:
        weight = float(clickedWeight.get())
        if weight <= 2 or weight >= 635:
            validation_errors.append("Weight must be from 2 to 635 kg.")
    except ValueError:
        validation_errors.append("Invalid weight input.")

    # Validate Height
    try:
        height = float(clickedHeight.get())
        if height < 55 or height > 272:
            validation_errors.append("Height must be from 55 to 272 cm.")
    except ValueError:
        validation_errors.append("Invalid height input.")

    # Validate Children
    try:
        children = int(clickedChildren.get())
        if children < children_min or children > children_max:
            validation_errors.append("Number of children must be between {} and {}.".format(children_min, children_max))
    except ValueError:
        validation_errors.append("Invalid children input.")

    if validation_errors:
        show_error_messages(validation_errors)
    else:
        show_Results()

def show_error_messages(errors):
    error_message = "\n".join(errors)
    messagebox.showerror("Validation Error", error_message)

mainFrame.mainloop()
```
